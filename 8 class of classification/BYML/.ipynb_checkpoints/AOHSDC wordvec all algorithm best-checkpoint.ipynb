```python
#Removing any date format 
df["Msg"].replace(
    {'\d+[\/:\-]\d+[\/:\-\s]*[\dAaPpMm]*' : '', '\w+\s\d+[\,]\s\d+' : ''},
    regex=True,
    inplace=True,
    )
```

```python
#function to remove non-ASCII 
def remove_non_ascii(text): 
    return ''.join(i for i in text if ord(i)<128) 
    df['Msg'] = df['Msg'].apply(remove_non_ascii)
```

```python
#
def tokenize_text(text):
    tokens = []
    for sent in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sent):
            if len(word) < 2:     # discard a word with 1 character 
                continue
            tokens.append(word)
    return tokens
# # basic tokenize 
```

```python
# Loading my custom word2ve
from gensim.models import KeyedVectors
model = KeyedVectors.load_word2vec_format('myword2vecfinal.bin', binary=True)
print(model)
```

```python
embeddingsSize=100
```

```python
import numpy as np
def getVectors(dataset):
  singleDataItemEmbedding=np.zeros(embeddingsSize)
  vectors=[]
  for dataItem in dataset:
    wordCount=0
    for word in dataItem:
      if word in model.vocab:
        singleDataItemEmbedding=singleDataItemEmbedding+model[word]
        wordCount=wordCount+1
       

    singleDataItemEmbedding=singleDataItemEmbedding/wordCount
    vectors.append(singleDataItemEmbedding)
  return vectors

trainReviewVectors=getVectors(X_train)
```

```python
trainReviewVectors
```

```python
M = np.concatenate([trainReviewVectors], axis=1) # use concatenate for multple featuer exteraction if any
M
```

```python
X_w2v = pd.DataFrame(M) # chkec M and N
y = df['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(trainReviewVectors,y,test_size=0.2,random_state=150)
```

# 3. Machine learnig algorithm 

# 3.1 SVM Algorithm

```python
#from sklearn.svm import SVC
from sklearn.svm import SVC
svmmodel=SVC(C=0.4, kernel='rbf',gamma='scale')
svmmodel.fit(X_train,y_train)
svmmodel.score(X_test, y_test)
```

```python
#by RF algorithm 
from sklearn.ensemble import RandomForestClassifier
# model train by Random forest 
rfcmodel=RandomForestClassifier(n_estimators=500,criterion='gini',max_features='log2', random_state = 42)
rfcmodel.fit(X_train,y_train)
rfcmodel.score(X_test, y_test)
```

```python
# minmax scale b/c multinominal Naive bayes not accept negative vectors 
from sklearn.preprocessing import MinMaxScaler
# transform data
scaler = MinMaxScaler()
xtrain = scaler.fit_transform(X_train)
xtest = scaler.fit_transform(X_test)
```

```python
# by multinominal  Naive Bayes M
# Naive Bayes Algorithm 1 by ngram 
#from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import MultinomialNB
nbmodel = MultinomialNB()
nbmodel.fit(xtrain,y_train)
nbmodel.score(xtest, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nbmodelg=GaussianNB()
nbmodelg.fit(X_train,y_train)
nbmodelg.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
nbmodelb=BernoulliNB()
nbmodelb.fit(X_train,y_train)
nbmodelb.score(X_test, y_test)
```

```python
# convert list to  numpy during data feading 
import numpy as np
X_train1 = np.array(X_train)
y_train1 = np.array(y_train)
X_test1 = np.array(X_test)
y_test1 = np.array(y_test)
```

```python
# Xgboost Algorithm 
import xgboost as xgb
xgbmodel=xgb.XGBClassifier(learning_rate=1.0, max_depth=2, n_estimators= 500)
xgbmodel.fit(X_train1,y_train1)
xgbmodel.score(X_test1, y_test1)
```
